```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv(r'C:\Users\joeym\Documents\Wake Forest\Pitching Lab\Baseball Analytics\advanced_pitching_stats.csv', sep=",")
df1
```

```python
df1 = df1.dropna()
```

```python
df1 = df1.loc[(df1['gp'] >= 5)]
df1
```

```python
dfconf = df1.loc[(df1['conference'] == 'CAA') | (df1['conference'] == 'Big Ten') | (df1['conference'] == 'Pac-12') | (df1['conference'] == 'Big 12') | (df1['conference'] == 'Sun Belt') | (df1['conference'] == 'SEC') | (df1['conference'] == 'ACC') | (df1['conference'] == 'ASUN') | (df1['conference'] == 'C-USA')]
dfconf
```

```python
fip = dfconf['fip'].groupby(dfconf['conference']).mean().sort_values(ascending=False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(fip)
```

```python
era = dfconf['era'].groupby(dfconf['conference']).mean().sort_values(ascending=False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(era)
```

```python
def WFipIp(dfconf): 
        return dfconf['fip'] * dfconf['ip']

dfconf['WFipIp'] = dfconf[['ip','fip']].apply(WFipIp, axis=1)

def WEraIp(dfconf): 
        return dfconf['era'] * dfconf['ip']

dfconf['WEraIp'] = dfconf[['ip','era']].apply(WEraIp, axis=1)
```

```python
dfconf['WFipSum'] = dfconf.groupby('name')['WFipIp'].transform('sum')
dfconf['WEraSum'] = dfconf.groupby('name')['WEraIp'].transform('sum')
dfconf['TotalIp'] = dfconf.groupby('name')['ip'].transform('sum')

def WFip(dfconf): 
    return dfconf['WFipSum'] / dfconf['TotalIp']

dfconf['WFip'] = dfconf[['WFipSum', 'TotalIp']].apply(WFip, axis=1)

def WEra(dfconf): 
    return dfconf['WEraSum'] / dfconf['TotalIp']

dfconf['WEra'] = dfconf[['WEraSum', 'TotalIp']].apply(WEra, axis=1)
dfconf
```

```python
fipplot = sns.lmplot('fip','WFip', data=dfconf)
fipplot.set(xlim=(0, 10))
fipplot.set(ylim=(0, 10))

dfconf[['fip','WFip']].corr()
```

```python
eraplot = sns.lmplot('era','WEra', data=dfconf)
eraplot.set(xlim=(0, 10))
eraplot.set(ylim=(0, 10))

dfconf[['era','WEra']].corr()
```

```python
dfconf['TeamIp'] = dfconf.groupby('team')['ip'].transform('sum')

def WFipSumAvg(dfconf): 
    return dfconf['WFipSum'] / 3

dfconf['WFipSumAvg'] = dfconf[['WFipSum']].apply(WFipSumAvg, axis=1)

dfconf['WFipSumTeam'] = dfconf.groupby('team')['WFipSumAvg'].transform('sum')

def WFipTeam(dfconf): 
    return dfconf['WFipSumTeam'] / dfconf['TeamIp']

dfconf['WFipTeam'] = dfconf[['WFipSumTeam', 'TeamIp']].apply(WFipTeam, axis=1)
dfconf
```

```python
def WEraSumAvg(dfconf): 
    return dfconf['WEraSum'] / 3

dfconf['WEraSumAvg'] = dfconf[['WEraSum']].apply(WEraSumAvg, axis=1)

dfconf['WEraSumTeam'] = dfconf.groupby('team')['WEraSumAvg'].transform('sum')

def WEraTeam(dfconf): 
    return dfconf['WEraSumTeam'] / dfconf['TeamIp']

dfconf['WEraTeam'] = dfconf[['WEraSumTeam', 'TeamIp']].apply(WEraTeam, axis=1)
dfconf
```

```python
dfconf.groupby('conference')['WFipTeam'].mean()
```

```python
dfconf.groupby('conference')['WEraTeam'].mean()
```

```python
dfconf['AvgFip'] = dfconf.groupby('name')['fip'].transform('mean')
dfconf['AvgK_Pct'] = dfconf.groupby('name')['k_pct'].transform('mean')
dfconf['AvgBB_Pct'] = dfconf.groupby('name')['bb_pct'].transform('mean')
dfconf['AvgHr9'] = dfconf.groupby('name')['hr9'].transform('mean')
dfconf
```

```python
def WFipTeamSos(dfconf): 
    
    if dfconf['conference'] == 'Sun Belt':
        return (dfconf['WFipTeam']) / 1.063961029
    elif dfconf['conference'] == 'C-USA':
         return (dfconf['WFipTeam']) / 1.046062387
    elif dfconf['conference'] == 'ASUN': 
         return (dfconf['WFipTeam']) / 1.027479452
    elif dfconf['conference'] == 'ACC': 
        return (dfconf['WFipTeam']) / 1.123815193
    elif dfconf['conference'] == 'CAA': 
        return (dfconf['WFipTeam']) / 1.023290149
    elif dfconf['conference'] == 'Big Ten': 
        return (dfconf['WFipTeam']) / 1.054370954
    elif dfconf['conference'] == 'Big 12': 
          return (dfconf['WFipTeam']) / 1.124006146
    elif dfconf['conference'] == 'Pac-12': 
        return (dfconf['WFipTeam']) / 1.089591957
    elif dfconf['conference'] == 'SEC': 
         return (dfconf['WFipTeam']) / 1.16431282
    else:
        return 'NA'
        
dfconf['WFipTeamSos'] = dfconf[['conference','WFipTeam']].apply(WFipTeamSos, axis=1)
```
