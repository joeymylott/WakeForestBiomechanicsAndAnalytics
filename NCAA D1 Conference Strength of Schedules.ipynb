```python
sostotal = pd.read_csv(r'C:\Users\joeym\Documents\Wake Forest\Baseball Analytics\Top Pitching Staff\NCAA D1 SOS.csv')
sostotal
```

```python
# Select every row with averageSOS from 2018-2020

avgsos = sostotal[sostotal.index % 3 == 2]
avgsos
```

```python
# Create average SOS for every conference

confsos = avgsos.groupby('Conference', as_index=False)['2018-2020 averageSOS'].mean()
confsos.columns = ['Conference','averageSOS']
confSOS = confsos.sort_values(by='averageSOS', ascending=False)
confSOS
```

```python
# Calculate relative SOS for each conference

meansos = confSOS['averageSOS'].mean()
relsos = np.divide(confSOS['averageSOS'], meansos)

# Merge data frames

sosdfs = [confSOS, relsos]

confrelsos = pd.concat(sosdfs, axis=1)
confrelsos.columns = ['Conference','averageSOS', 'relSOS']
confrelSOS = confrelsos.sort_values(by='relSOS', ascending=False)
confrelSOS
```
